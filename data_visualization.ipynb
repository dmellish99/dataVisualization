```
import pandas as pd

# Cargar archivos CSV
tracks = pd.read_csv('track_info.csv')
sections = pd.read_csv('sections.csv')
artists = pd.read_csv('artists.csv')
segments = pd.read_csv('segments.csv')
```

```
# Información esencial de cada CSV
print("Información del CSV de tracks:")
tracks.info()

print("\nInformación del CSV de sections:")
sections.info()

print("\nInformación del CSV de artists:")
artists.info()

print("\nInformación del CSV de segments:")
segments.info()
```

```
# Ver valores faltantes
print("Valores faltantes en el CSV de tracks:")
print(tracks.isnull().sum())

print("\nValores faltantes en el CSV de sections:")
print(sections.isnull().sum())

print("\nValores faltantes en el CSV de artists:")
print(artists.isnull().sum())

print("\nValores faltantes en el CSV de segments:")
print(segments.isnull().sum())
```

```
# Estadísticas básicas
print("Estadísticas del CSV de tracks:")
print(tracks.describe())

print("\nEstadísticas del CSV de sections:")
print(sections.describe())

print("\nEstadísticas del CSV de artists:")
print(artists.describe())

print("\nEstadísticas del CSV de segments:")
print(segments.describe())
```
